### 파이썬 알고리즘 공부 선택정렬

#### 선택정렬의 기본적인 아이디어는 바로
#### "가장 작은 것을 제일 앞으로 보내면 어떨까?"

```python
my_list = [1, 10, 5, 8, 7, 6, 4, 3, 2, 9]
```

#### 앞에서부터 차례대로 숫자를 비교한 후 값이 같은 경우에 숫자를 서로 변경한다.

```python
#내가 생각한 코드

for i in range(10):
    for j in range(10):
        if i < j:
            if my_list[i] > my_list[j]:
                my_list[i], my_list[j] = my_list[j], my_list[i]
                print("중간결과: ",my_list)
print(my_list)
        
       
           
```

#### i와 j를 비교하지않고 해도 진행가능하다.

```python
my_list = [1, 10, 5, 8, 7, 6, 4, 3, 2, 9]
```

### 일반적인 방법 1

```python
n = len(my_list)

for i in range(n-1):
    for j in range(i+1, n):
        if my_list[i] > my_list[j]:
                my_list[i], my_list[j] = my_list[j], my_list[i]
                print("중간결과: ",my_list)
print(my_list)
        
```

### 선택정렬 알고리즘의 경우 각 값을 10+ 9+ 8+ 7+...+1 = 55번의 n(n+1)/2 = 55

### 수행시간은 n(n+1)/2  이지만 n이 엄청 크다고 생각하면 숫자로 나누고 더하는 과정은 의미가 없다.
### 따라서 수행시간은 n*n 으로 표현 할 수 있다.

### O(N * N)로 표기할 수 있음( 처리시간 표기 법)

#### 데이터 개수가 만개면 1억번정도의 처리를 해야함


## 선택정렬의 처리시간은 O(N^2)
#### 처리데이터양이 많은경우 가장 피해야할 알고리즘. 데이터양이 증가하면 처리시간은 기하급수적으로 늘어난다
